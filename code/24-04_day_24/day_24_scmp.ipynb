**China's High-Speed Rail Boom (Modern Fantasy / Dark Edition)**  
Reimagining the SCMP infographic style with a modern, futuristic, neon-fantasy aesthetic. It retains the classic SCMP layout (tall aspect ratio, negative-space annotations) but translates it into a striking dark mode with glowing elements, representing the high-tech reality of China's HSR network.

**Data:** China State Railway Group / Official Government Data  
**Author:** Sharfudeen Yasar Arafath

```r
library(ggplot2)
library(dplyr)
library(showtext)
library(sysfonts)
library(scales)
```

```r
# Modern, futuristic fonts
font_add_google("Outfit", "outfit")
font_add_google("JetBrains Mono", "jetbrains")
font_add_google("Roboto Condensed", "roboto_condensed")
showtext_auto()
showtext_opts(dpi = 300)

# Using a tall "infographic" aspect ratio
options(repr.plot.width = 10, repr.plot.height = 12, repr.plot.res = 300)
```

```r
df <- read.csv("../../data/day_24/china_hsr.csv")

# Modern Fantasy / Dark Cyber Aesthetics
bg_col     <- "#0b0f19"  # Deep void/space background
text_col   <- "#f8fafc"  # Bright white for primary text
neon_red   <- "#ff003c"  # Glowing cyber red
neon_pink  <- "#ff4d85"  # Lighter pink for accents and core lines
grid_col   <- "#1e293b"  # Dark subtle grid
accent_col <- "#94a3b8"  # Slate grey for secondary text
```

```r
p <- ggplot(df, aes(x = year, y = length_km)) +
  # Fantasy/Sci-Fi Grid (Horizontal only)
  geom_hline(yintercept = seq(10000, 50000, 10000), colour = grid_col, linewidth = 0.5, linetype = "dashed") +
  
  # Glowing Area Effect (Faint neon wash)
  geom_area(fill = neon_red, alpha = 0.1) +
  
  # Glowing Line Effect (Layered for neon glow)
  geom_line(colour = neon_red, linewidth = 4, alpha = 0.3) +   # Outer glow
  geom_line(colour = neon_pink, linewidth = 1.5) +             # Core line
  geom_line(colour = "#ffffff", linewidth = 0.5, alpha = 0.9) + # Ultra-hot core
  
  # 2008 Annotation (Digital HUD style)
  annotate("text", x = 2008.2, y = 14000, label = "2008\nFirst 350km/h\nline opens\n(Beijing-Tianjin)", 
           family = "jetbrains", size = 4, hjust = 0, colour = text_col, lineheight = 0.9) +
  annotate("segment", x = 2008.5, xend = 2008, y = 10000, yend = 2000, colour = accent_col, linewidth = 0.5, linetype = "dotted") +
  annotate("point", x = 2008, y = 1042, colour = neon_pink, size = 3.5, shape = 21, fill = bg_col, stroke = 1.5) +
  
  # 2013 Annotation
  annotate("text", x = 2011, y = 22000, label = "2013\nNetwork surpasses\n10,000 km", 
           family = "jetbrains", size = 4, hjust = 0, colour = text_col, lineheight = 0.9) +
  annotate("segment", x = 2012, xend = 2013, y = 18500, yend = 12000, colour = accent_col, linewidth = 0.5, linetype = "dotted") +
  annotate("point", x = 2013, y = 11028, colour = text_col, size = 3, shape = 21, fill = bg_col, stroke = 1.2) +
  
  # 2016 Annotation
  annotate("text", x = 2014, y = 32000, label = "2016\nSurpasses\n20,000 km", 
           family = "jetbrains", size = 4, hjust = 0, colour = text_col, lineheight = 0.9) +
  annotate("segment", x = 2015, xend = 2016, y = 29000, yend = 23500, colour = accent_col, linewidth = 0.5, linetype = "dotted") +
  annotate("point", x = 2016, y = 22669, colour = text_col, size = 3, shape = 21, fill = bg_col, stroke = 1.2) +
  
  # 2021 Annotation
  annotate("text", x = 2018, y = 45000, label = "2021\nSurpasses\n40,000 km", 
           family = "jetbrains", size = 4, hjust = 0, colour = text_col, lineheight = 0.9) +
  annotate("segment", x = 2019.5, xend = 2021, y = 43000, yend = 41000, colour = accent_col, linewidth = 0.5, linetype = "dotted") +
  annotate("point", x = 2021, y = 40000, colour = text_col, size = 3, shape = 21, fill = bg_col, stroke = 1.2) +
  
  # 2025 Target/Milestone
  annotate("text", x = 2024.8, y = 53500, label = "2025\n50,000 km\nmilestone reached", 
           family = "jetbrains", size = 4.5, hjust = 1, colour = neon_pink, fontface = "bold", lineheight = 0.9) +
  annotate("point", x = 2025, y = 50000, colour = neon_red, size = 6, alpha = 0.4) +
  annotate("point", x = 2025, y = 50000, colour = neon_pink, size = 3.5, shape = 21, fill = bg_col, stroke = 1.5) +
  
  # Axes
  scale_y_continuous(breaks = seq(0, 50000, 10000), 
                     labels = scales::comma,
                     expand = expansion(mult = c(0, 0.12))) +
  scale_x_continuous(breaks = seq(2008, 2024, 4),
                     expand = expansion(mult = c(0.02, 0.05))) +
  
  labs(
    title = "China's High-Speed Rail Boom",
    subtitle = "In less than two decades, China built the world's largest high-speed rail network,\nreaching the 50,000 km milestone in 2025.",
    caption = "Source: China State Railway Group / Official Government Data\nGraphic: Sharfudeen Yasar Arafath | #30DayChartChallenge \u00b7 Day 24",
    x = NULL, 
    y = "OPERATING LENGTH (KM)"
  ) +
  
  # Theme
  theme_minimal(base_family = "outfit", base_size = 14) +
  theme(
    plot.background    = element_rect(fill = bg_col, colour = NA),
    panel.background   = element_rect(fill = bg_col, colour = NA),
    panel.grid.major.x = element_blank(),
    panel.grid.minor   = element_blank(),
    panel.grid.major.y = element_blank(),
    
    plot.title         = element_text(family = "outfit", face = "bold", size = 28, colour = text_col, margin = margin(b = 10)),
    plot.subtitle      = element_text(family = "roboto_condensed", size = 14, colour = accent_col, lineheight = 1.1, margin = margin(b = 40)),
    plot.caption       = element_text(family = "jetbrains", size = 9, colour = accent_col, hjust = 0, lineheight = 1.4, margin = margin(t = 30)),
    
    axis.text.x        = element_text(family = "jetbrains", size = 12, colour = accent_col, margin = margin(t = 8)),
    axis.text.y        = element_text(family = "jetbrains", size = 12, colour = accent_col, margin = margin(r = 5)),
    axis.title.y       = element_text(family = "jetbrains", size = 12, colour = accent_col, hjust = 1, angle = 90, margin = margin(r = 10)),
    
    plot.margin        = margin(40, 50, 40, 50)
  )

p
```

```r
ggsave("../../chart/day_24_scmp.png", plot = p,
       width = 10, height = 12, dpi = 300, bg = bg_col)
cat("\u2705 SCMP Modern Fantasy Chart saved\n")
```
